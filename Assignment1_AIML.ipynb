```python
import numpy as np
import pandas as pd
```

```python
list1=[1,2,3,4,5]
list2=[5,10,15,20,25]
array1 = np.array(list1)
array2 = np.array(list2)

print("list1:", type(list1))
print("array1:", type(array1))
```

```python
#multiply
multiplication= array1 * array2
print("multiplication is:",multiplication)
```

```python
#divion
division= array1 / array2
print("division is:",division)
```

```python
#poweroff
poweroff=array1 ** 2
print("poweroff of array1 is:",poweroff)

poweroff=array2 ** 3
print("poweroff of array2 is:",poweroff)
```

```python
#addition of two
addition= array1 + array2
print("addition of two:",addition)
```

```python
sin_value=np.sin(array1)
log_value=np.log(array2)
log2_value=np.log2(array1)
exp_value=np.exp(array2)

print("sin value:", sin_value)
print("log value:",log_value)
print("log2 value:",log2_value)
print("exp value:",exp_value)
```

```python
import matplotlib.pyplot as plt
```

```python
# Given information
people = ['Ravi', 'Raju', 'Abhi', 'Arjun']
age = [21, 27, 41, 37]
height = [163, 149, 158, 171]
weight = [62, 75, 72, 69]
```

```python
#scatter plot
plt.figure(figsize=(7,9))
plt.scatter(age, height,s = weight,c='Red',alpha=0.7)

# Adding labels and title
plt.title('Scatter Plot - Age vs Height (Weight represented by size)')
plt.xlabel('Age')
plt.ylabel('Height')
# Show the plot
plt.show()
```

```python
data = {'PEOPLE': ['Ravi', 'Raju', 'Abhi', 'Arjun'],
        'AGE': [21, 27, 41, 37],
        'HEIGHT': [163, 149, 158, 171],
        'WEIGHT': [62, 75, 72, 69]}
```

```python
df = pd.DataFrame(data)
```

```python
# Bar chart
plt.bar(df['PEOPLE'], df['WEIGHT'], color='yellow')
plt.title('Bar Chart of Weight for People')
plt.xlabel('People')
plt.ylabel('Weight')
plt.show()
```

```python
df = pd.DataFrame(data)
```

```python
# Histogram for Age
plt.hist(df['AGE'], bins=10, color='Green', edgecolor='black')
plt.title('Histogram of Age for People')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Histogram for Height
plt.hist(df['HEIGHT'], bins=10, color='orange', edgecolor='black')
plt.title('Histogram of Height for People')
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.show()

# Histogram for Weight
plt.hist(df['WEIGHT'], bins=10, color='black', edgecolor='black')
plt.title('Histogram of Weight for People')
plt.xlabel('Weight')
plt.ylabel('Frequency')
plt.show()
```
